```python
def calculate_loss(self, y_true, y_pred):
        m = y_true.shape[0]
        # Add a tiny epsilon to avoid log(0) errors
        epsilon = 1e-15
        y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
        
        loss = -1/m * np.sum(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
        return loss
```
